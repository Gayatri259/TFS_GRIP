#**Importing** **the** **Libraries**

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#**Reading** **the** **Dataset**

```
df=pd.read_csv("globalterrorismdb_0718dist.csv", encoding='latin1')
df
```

```
df.describe()
```

```
df.isnull().sum()
```

```
df.shape
```

#**Relationship** **Amongst** **Various** **Components** **of** **the** **Dataset**

```
df.hist(figsize=(30,35))
plt.show
```

#**Countries** **that** **got** **Attacked** **the** **Most**

```
df['country'].value_counts()
plt.figure(figsize=(10,5))
sns.barplot(x=df['country_txt'].value_counts()[:20].index, y=df['country_txt'].value_counts()[:20].values)
plt.title('Countries that got Attacked the Most')
plt.xlabel('Countries')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=90)
plt.show()
```

###Iraq is the country that got attacked the most followed by Pakistan.

#**Terrorist** **Attacks** **in** **Various** **Regions**

```
df['region_txt'].value_counts()
plt.figure(figsize=(10,5))
sns.countplot(x='region_txt', data=df)
plt.title('Terrorist Attacks in Various Regions')
plt.xlabel('Regions')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=90)
plt.show()
```

### Middle East and North Africa was the most attacked region followed by South Asia.

# **Terror** **Attacks** **(1970-2017)**

```
plt.subplots(figsize=(10,5))
sns.countplot(x='iyear', data=df)
plt.title('Terror Attacks (1970-2017) ')
plt.xlabel('Years')
plt.ylabel('Number of Attacks')
plt.show()
```

### We see a huge rise in the terrorist attacks from 2008 to 2014.

# **Types** **of** **Attacks**

```
plt.figure(figsize=(10,5))
sns.countplot(x=df['attacktype1_txt'], order=df['attacktype1_txt'].value_counts().index)
plt.xlabel('Type of Attacks')
plt.ylabel('Number of Attacks')
plt.title('Types of Attacks')
plt.xticks(rotation=90)
plt.show()
```

### Bombing/Explosions are the most commonly used types of attacks followed by Armed Assault.

# **Weapons** **Used** **During** **Attacks**

```
df['weaptype1_txt'].value_counts()
plt.figure(figsize=(10,5))
sns.barplot(x=df['weaptype1_txt'].value_counts()[:20].index, y=df['weaptype1_txt'].value_counts()[:20].values)
plt.title('Weapons Used During Attacks')
plt.xlabel('Weapons')
plt.ylabel('Number of Times')
plt.xticks(rotation=90)
plt.show()
```

### Explosives are the most commonly used weapons followed by Firearms.

# **Targets** **of** **Attacks**

```
plt.figure(figsize=(10,5))
sns.countplot(x=df['targtype1_txt'], order=df['targtype1_txt'].value_counts().index)
plt.title('Targets of Attacks')
plt.xlabel('Targets')
plt.ylabel('Number of Targets')
plt.xticks(rotation=90)
plt.show()
```

### Private Citizens and Property are targeted the most by terrorists followed by Military.

# **Names** **of** **Groups** **that** **Attacked**

```
plt.figure(figsize=(15,5))
sns.barplot(x=df['gname'].value_counts()[:20].index, y=df['gname'].value_counts()[:20].values)
plt.title('Names of Groups that Attacked')
plt.xlabel('Name of the Group')
plt.ylabel('Total Number of Attacks')
plt.xticks(rotation=90)
plt.show()
```

### Most attacks are done by Unknown groups followed by Taliban.

# **Cities** **that** **got** **Affected** **by** **Terrorist** **Attacks**





```
df['city'].value_counts()
```

```
df['city'].value_counts()
plt.figure(figsize=(10,5))
sns.barplot(x=df['city'].value_counts()[:20].index, y=df['city'].value_counts()[:20].values)
plt.title('Cities that got the most Affected by Terrorist Attacks')
plt.xlabel('Cities')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=90)
plt.show()
```
